```python
#30. Python | Program to print duplicates from a list of integers

list=[1,2,3,4,5,6,22,33,3,-20,22]
 
uniqueList = []
duplicateList = []
 
for i in list:
    if i not in uniqueList:
        uniqueList.append(i)
    elif i not in duplicateList:
        duplicateList.append(i)
 
print(duplicateList)
```
