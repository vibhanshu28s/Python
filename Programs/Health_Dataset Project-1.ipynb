```python
#Replace using mean, median or mod.
#Mean
df=pd.read_csv("/Users/Vibhanshu/Desktop/Health_data.csv")
x=df["Calories"].mean()
df["Calories"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#Replace using mean, median or mode.
#Median
df=pd.read_csv("/Users/Vibhanshu/Desktop/Health_data.csv")
x=df["Calories"].median()
df["Calories"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#Mode
df=pd.read_csv("/Users/Vibhanshu/Desktop/Health_data.csv")
x=df["Calories"].mode()[0]
df["Calories"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#Wrong Data
df=pd.read_csv("/Users/Vibhanshu/Desktop/Health_data.csv")
df.loc[7,"Duration"]=45
print(df.to_string())
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.loc[x,"Duration"]=120
print(df.to_string())        
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.drop(x,inplace=True)
print(df.to_string())   
```
