```python
#Draw a line in a diagram from position(0,0)to(6,250).
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid()
plt.show()
```

```python
#Labels
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
#Subplot
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
#Bar chart -vertical
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
#Bar chart -horizontal
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="green")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="green",width=0.1)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="green",height=0.1)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 100000)

plt.hist(x, 100)
plt.show()
```

```python
#Pie Chart
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
plt.pie(y,labels=mylable1)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
plt.pie(y,labels=mylable1,startangle=90)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylable1,explode=myexplode)

plt.show()
```

```python
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylable1,explode=myexplode,shadow=True)

plt.show()
```

```python
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
mycolor=["red","y","hotpink","g"]
plt.pie(y,labels=mylable1,colors=mycolor)
plt.show()
```

```python
#Legend()
#Pie Chart
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
plt.pie(y,labels=mylable1)
plt.legend()
plt.show()
```

```python
#legend with heder
#Legend()
#Pie Chart
y=np.array([35,25,25,15])
mylable1=["Apple","Banana","Cherry","Grapes"]
plt.pie(y,labels=mylable1)
plt.legend(title="Fruits")
plt.show()
```
